```python
data = [1, 7, 11, 12, 14, 23, 44, 47, 51, 64, 67, 77,140, 672, 871]

def binary_search(data, target):
    s = 0
    e = len(data) -1

    while s <= e:
        m = (s + e)//2
        if target == data[m] :
            return m
        elif target > data[m]:
            s = m + 1
        else:
            e = m -1
    else:
        return ' '
        
print(binary_search(data,1))
print(binary_search(data,871))
```

```python
import numpy as np

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

targets = np.array([[0],[0],[0],[1]])

def step_function(a):
    
     return np.array( a>= 0, dtype=np.int)
    
def pcntrain(inputs, targets, l_rate, cnt):
    
    W = np.array([[0.3], [0.4], [0.1]])
    x0 = -1
    
    for i in range(cnt):
        for j,x in enumerate(inputs):
            k = x0 * W[0][0] + x[0] * W[1][0] + x[1] * W[2][0]
            cost = targets[j] - step_function(k)
            if cost!= 0:
                W = change_weight(W, l_rate, x0, x, cost) 
    
    for i, w in enumerate(W):
        print("w"+str(i) + " = "+ str(w[0]))
                
def change_weight(W, l_rate, x0, x, cost):
    
    X = [x0] + list(x)
    for i, w in enumerate(W):
        W[i] = w + l_rate * X[i] * cost
        
    return W

pcntrain(inputs,targets,0.05,10)
```

```python
import copy

def friends_block(m,n,board, N):
    #   :     ->  pop icon 
    while True:
        #  : 9 
        b = []
        for i in range(m-(N-1)):
            for j in range(n-(N-1)):
                if board[i][j] != ' ':
                    #  -> 
                    a = make_block(N,i,j)
                    c = []
                    for k, l in a:
                        if board[i][j] == board[k][l]:
                            c.append(1)
                    
                    if c.count(1) == N**2:
                        b = b + a.copy()
        # end option
        if len(b) == 0:
            break
            
        # 4    
        for i, j in b:
            board[i] = board[i][:j] + ' ' + board[i][j+1:]

        board = pull_down(board,m,n)
        
    # 
    c = 0
    for i in range(m):
        for j in range(n):
            if board[i][j] == ' ':
                c +=1
    return c

def make_block(n,k,l):
    a = []
    i, j = 1, 0
    for i in range(n):
        for j in range(n):
            a.append((i+k,j+l))
    return a

#  ! -> 
def pull_down(board,m,n):
    for i in range(m):
        for j in range(n):
            if board[i][j] == ' ':
                if i != 0:
                    board[i] = board[i][:j] + board[i-1][j] + board[i][j+1:]
                    board[i-1] = board[i-1][:j] + ' ' + board[i-1][j+1:]

    return board

# # :, : 
# m,n= map(int,input().split())
# board = eval(input())
m, n = 6, 6
board = ["TTTANT", "RRRACC", "RRRRCC", "RRRRAA", "TTTMMF", "TTTTTJ"]
m1,n1 = 6,6
board2 = ["TTTANT", "RRFACC", "RRRFCC", "TRRRAA", "TTMMMF", "TMMTTJ"]
m2, n2 = 8,8
board3 = ["AAABBBCC", "AAAAACCC", "AAAAACCC", "TTAAAAAA", "BBBBBBCC", "AAABBBAA", "AAABBBAA", "AAACCCAA"]
print(friends_block(m,n, board,3))
print(friends_block(n1,m1, board2, 2))
print(friends_block(n2,m2, board3, 3))
print(friends_block(n2,m2, board3, 2))
print(friends_block(n2,m2, board3, 5))
```
