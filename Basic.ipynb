    1. 숫자
    
    2. 문자

        예 : t = (1,2,3)

             m

        예 : s1 = set([1,2,3,4,5,6])
             s2 = set([4,5,6,7,8,9])
             
             print(s1 & s2) # 교집합
             print(s1 | s2) # 합집합
             print(s1 - s2) # 차집합
        

```python
l = [1,2,3]
l[0] = 5
print(l)

t = (1,2,3)
# t[0] = 5
print(t)

m = {}
m['python'] = 'www.python.org'
print(m)

s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

print(s1 & s2) # 교집합
print(s1 | s2) # 합집합
print(s1 - s2) # 차집합
```

■ 10. if문 개념 배우기
    
    2. if - elif

```python
# 예: 이름을 물어보게 하고 이름을 입력하면 해당 사원의 월급이 
# 3000 이상이면 고소득자 2000 ~ 3000 사이이면 적당합니다. 2000 이하이면 월급 인상자입니다. 라는 메세지가 출력되게 하시오

name = input('이름을 입력하세요 : ')

import csv
file = open("emp2.csv", 'r')

emp_csv = csv.reader(file)

for emp_list in emp_csv:
    if emp_list[1] == name.upper():
        if int(emp_list[5]) >= 3000:
            print('고소득자 입니다.')
        elif int(emp_list[5]) >= 2000:
            print('적당합니다.')
        else:
            print('월급 인상대상자 입니다.')
```

```python
# 문제12. 이름을 물어보게하고 이름을 입력했을 때
# 해당 사원의 직업이 SALESMAN이면 부서이동 대상자입니다.
# ANALYST이면 지역이동 대상자입니다.
# 나머지 직업은 해당사항 없습니다. 라는 메세지를 출력하시오

name = input('이름을 입력하세요 : ')

import csv
file = open("emp2.csv", 'r')

emp_csv = csv.reader(file)

for emp_list in emp_csv:
    if emp_list[1] == name.upper():
        if emp_list[2] == 'SALESMAN':
            print('부서이동 대상자입니다.')
        elif emp_list[2] == 'ANALYST':
            print('지역이동 대상자입니다.')
        else:
            print('대상자가 아닙니다.')
```

■ 12. for 문

        1. continue 문
            " 반복문이 실행되는 동안 특정 코드 블럭을 실행하지 않고 다른 코드 블럭만 실행되게 할 때 사용하는 문법


```python
# 예 : 0 부터 9까지 출력하는데 홀수는 출력하지 마라
    
for i in range(10):
    if i % 2 == 1:
        continue
    print(i)
```

■ 37. 튜플 이해하기 (())

튜플(tupple) 변수란?

    리스트 변수와는 다르게 요소 데이터의 변경이 불가능한 자료형 리스트는 대괄호 [ ] 로 요소들을 둘렀는데, 튜플은 소괄호 ( ) 로 요소를 감싼다
    " 변경이 안되므로 튜플로 만든 데이터에 대한 신뢰도가 높아진다. "


```python
# 예제 : 
tuple1 = (1, 2, 3, 4, 5)
tuple2 = ('a', 'b', 'c')
tuple3 = (1, 'a', 'abc', [1, 2, 3, 4, 5], ['a', 'b', 'c'])
#tuple1[0] = 6

def myfunc():
    print('안녕하세요')

tuple4 = (1, 2, myfunc)   # 함수도 튜플에 속할 수 있음
tuple4[2]()       # ‘안녕하세요’ 가 출력됨
```
