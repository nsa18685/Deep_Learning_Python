```python
y_train_9  = (y_train == 9) # 9:True, others:False <- 이게 label?
y_test_9 = (y_test == 9)
```

```python
from sklearn.linear_model.stochastic_gradient import SGDClassifier  # --> SVM 중 linear 사용
sgd_clf = SGDClassifier(random_state=42, max_iter=50)
# sgd_clf = SGDClassifier(random_state=42, max_iter=50, verbose=1) # 확률적 경사 하강법: 무작위성을 사용, max_iter: 샘플을 반복함
sgd_clf.fit(X_train, y_train_9)
sgd_clf.predict([some_digit]) # ?
```

#### 성능측정

```python
# 교차 검증 구현
from sklearn.model_selection import StratifiedKFold  # 비율유지하도록 계층적 샘플링
from sklearn.base import clone # 새로운 추정량을 산출

skfolds = StratifiedKFold(n_splits=9, random_state=42)   # 10등분 이상으로 나누는 것은 거의... with 하이퍼파라미타 조절
for train_index, test_index in skfolds.split(X_train, y_train_9):
    clone_clf = clone(sgd_clf) # Constructs a new estimator with the same parameters.
    X_train_folds = X_train[train_index]
    y_train_folds = (y_train_9[train_index])
    X_test_fold = X_train[test_index]
    y_test_fold = (y_train_9[test_index])
    clone_clf.fit(X_train_folds, y_train_folds)
    
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred)) 
    
    
# bootstraping: 복원추출
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_9, cv=3, scoring="accuracy")
# but 9:[1:8] --> 불균형
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.xlabel("Threshold")
    plt.legend(loc="upper left")
    plt.ylim([0, 1])
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

#### The ROC Curve

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_9, y_scores)

def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
plot_roc_curve(fpr, tpr)
plt.show()
```

#### Error Analysis 
-> 분석후 없애는 작업이 중요!!

```python
y_train_pred = cross_val_predict(sgd_clf, X_train_scaled, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)
conf_mx
```

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()
# 1이 w여서 더 잘 되어 보이지만 성분이 더 많아서 그래 보일 수 있음
```

```python
# 대각선 성분을 없애고 정규화함 
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
```

```python
# 대각선 성분을 없애고 정규화함 -> black
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
# idea1: 8-9가 전체적으로 밝음: 잘 못맞춘편 --> 시간을 더 투자해야함.
# idea2: 1은 어두움: 잘 맞춘편
# idea3: 5:3, 3:5 대칭적 --> 비슷하게 생김
```

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
```

```python
# idea3: 5:3, 3:5 대칭적 --> 비슷하게 생김 --> 더 잘분리할 수 있도록 하자
cl_a, cl_b = 3, 5
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]
plt.figure(figsize=(8,8))
plt.subplot(221); plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222); plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223); plot_digits(X_ba[:25], images_per_row=5)
plt.subplot(224); plot_digits(X_bb[:25], images_per_row=5)
plt.show()
```
