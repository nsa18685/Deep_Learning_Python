```python
# 문제100. instr 함수를 파이썬으로 구현하시오

def instr(data, w):
    for i in range(len(data)):
        if data[i] == w :
            return i+1
    return 'null'

print(instr('smith','m'))
print(instr('smith','a'))
```

※ 아래 두 날짜 사이의 개월 수?

2017 - 01 - 01 ~ 2018 - 07- 01

2018 - 2017 = 1년 --> 12개월
07 - 01 = 06 --> 6개월

```python
# 문제116. months_between 함수를 파이썬으로 구현하시오

def months_between(date1, date2):
    import datetime

    dt1 = datetime.datetime.strptime(date1,'%Y-%m-%d')
    dt2 = datetime.datetime.strptime(date2,'%Y-%m-%d')
    
    
    year_b = abs(dt1.year - dt2.year)
    if dt1.month > dt2.month: 
        month_b = year_b*12 + (dt1.month - dt2.month)
    else:
        month_b = year_b*12 - (dt1.month - dt2.month)
        
    return month_b

print(months_between('2017-01-01','2018-07-01'))
print(months_between('2018-07-01','2017-01-01'))
```

■ 41. 지역변수와 전역변수 이해하기 (global)

   지역변수 :" 변수는 자신이 생성된 범위(코드블럭) 안에서만 유효하다.
               함수 안에서 만든 변수는 함수 안에서 살아 있다가 함수코드의 실행이 종료되면 그 생명을 다한다. "
               
   전역변수 : " 함수 외부에서 만든 변수는 프로그램이 살아있는 동안에 함께 살아있다가 프로그램이 종료되면 같이 소멸된다.
                프로그램 전체를 유효범위로 가진다. "

```python
# 예제:
param = 10

strdata = '전역변수'



def func1():

    strdata = '지역변수'

    print(strdata)


def func2(param):

    param = 1   # local param
# 함수 안에서 만든 변수는 함수 안에서 살아 있다가 함수코드의 실행이 종료되면 그 생명을 다한다.

def func3():

    global param

    param = 50


func1()                  # ‘지역변수’가 출력됨

print(strdata)           # ‘전역변수’가 출력됨

print(param)             # 10이 출력됨

func2(param)

print(param)             # 10이 출력됨

func3()

print(param)             # 50이 출력됨
```

```python
# 문제185. 파이썬은 짝수를 좋아한다. 2.5가 반올림해서 짝수가 되려면 어떻게 해야하나?

import math
a = 2.25
print(math.ceil(a))

# round( a + 0.5 ) 를 사용해서 올림의 효과를 만들어 냅니다.
print(a+0.5)

from decimal import Decimal,ROUND_HALF_UP
print(Decimal(2.56).quantize(0,ROUND_HALF_UP))
```

```python
# 문제186. 문제184 번을 다시 출력하는데 일의 자리에서 반올림되도록 하시오

emp_c['sal'] = emp['sal']*0.15 + 0.05
emp_c['sal'] = emp_c['sal'].round(-1)
print(emp_c[['ename', 'sal']])
```

■ 정수 리스트에서 소수만 걸러내기 (filter)

```python
# 예:
def getPrime(x):
    # for / else 구문
    for i in range(2, x-1):
        if x%i == 0:
            break

    else:
        return x


listdata = [117, 119, 1113, 11113, 11119]

ret = filter(getPrime, listdata)  # 하나씩 실행
# ret2 = getPrime(listdata)  # Error 발생

print(list(ret))
```

■ 최대값, 최소값 구하기 (min, max)

```python
listdata = [9.96, 1.27, 5.07, 6.45, 8.38, 9.29, 4.93, 7.73, 3.71, 0.93]

maxval = max(listdata)

minval = min(listdata)

print(maxval)     # 9.96이 출력됨

print(minval)     # 0.93이 출력됨



txt = 'Alotofthingsoccureachday'

maxval = max(txt)

minval = min(txt)

print(maxval);     # ‘y’가 출력됨

print(minval)     # ‘A’가 출력됨



maxval = max(2+3, 2*3, 2**3, 3**2)

minval = min('abz', 'a12')

print(maxval)

print(minval)
```
