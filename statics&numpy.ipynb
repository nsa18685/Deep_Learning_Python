※ 사분위수 범위란 무엇인가?

사분위수범위는 전체 데이터 중 가운데 부분인 50% 데이터의 산포 정도를 측정하는 통계량이다

범위 = 최대값 - 최소값
사분위수 범위 = 3사분위수 - 1사분위수




```python
# 문제24. 전체 범위를 구하는 함수를 파이썬으로 구현하시오
def data_range(x):
    return max(x) - min(x)

print(data_range(x))
```

※ 분산이란?

※ 표준편차란?

    분산의 양의 제곱근으로 원시 데이터의 측정 단위가 같은 단위로 퍼짐 정도를 나타내는 것

```python
# 문제28. 평균을 구하는 함수를 파이썬으로 구현하시오

import numpy as np

x = [0.062, 0.071, 0.067, 0.068, 0.066, 0.062, 0.068, 0.067, 0.060, 0.065]

def mean(args):
    return sum(args) / len(args)

print(mean(x))
```

```python
# 문제29. 분산을 구하는 함수를 파이썬으로 구현하시오

x = [0.062, 0.071, 0.067, 0.068, 0.066, 0.062, 0.068, 0.067, 0.060, 0.065]

def variance(args):
    re = 0
    m = mean(args)
    n = len(args)
    for i in args:
        re+= (i - m) ** 2
        
    return re/(n-1)

print(round(variance(x),5))
```

```python
# 문제30. 표준편차를 구하는 함수를 파이썬으로 구현하시오
import math

def stddev(args):
    return math.sqrt(variance(args))

print(stddev(x))
```
