※ 표준편차란?

    분산의 양의 제곱근으로 원시 데이터의 측정 단위가 같은 단위로 퍼짐 정도를 나타내는 것

```python
# 문제28. 평균을 구하는 함수를 파이썬으로 구현하시오

import numpy as np

x = [0.062, 0.071, 0.067, 0.068, 0.066, 0.062, 0.068, 0.067, 0.060, 0.065]

def mean(args):
    return sum(args) / len(args)

print(mean(x))
```

```python
# 문제29. 분산을 구하는 함수를 파이썬으로 구현하시오

x = [0.062, 0.071, 0.067, 0.068, 0.066, 0.062, 0.068, 0.067, 0.060, 0.065]

def variance(args):
    re = 0
    m = mean(args)
    n = len(args)
    for i in args:
        re+= (i - m) ** 2
        
    return re/(n-1)

print(round(variance(x),5))
```

```python
# 문제30. 표준편차를 구하는 함수를 파이썬으로 구현하시오
import math

def stddev(args):
    return math.sqrt(variance(args))

print(stddev(x))
```

```python
# 문제33. 네 사람 A,B,C,D 를 나란히 있는 네개의 의자에 배치시키려고 한다.
# 네 사람을 배치시키는 전체의 경우의 수와 그중 A가 가장 왼쪽에 배치될 경우의 수를 구하여라.
# A가 가장 왼쪽에 배치되는 확률은 얼마인가 ? 
# 전체 경우의 수 4! A가 가장 왼쪽에 배치될 경우의 수 3!
# A가 가장 왼쪽에 배치되는 확률 =  A가 가장 왼쪽에 배치되는 경우의 수 / 전체 경우의 수 = 3! / 4!

import math
print(math.factorial(4))
print(math.factorial(3))
print(math.factorial(3) / (math.factorial(4)))
```

```python
# 문제41. 한 동네에 200가구가 살고 있다. 각구가 지난 1년동안 병원에서 찾은 횟수를 조사해보니 다음과 같다.
# 확률변수를 X = '병원방문횟수' 로 하여 확률 분포함수와 누적확률 분포 함수를 아래와 같이 파이썬으로 구현하시오

def h_function(li):
    H = []
    for i in li:
        H.append(i/sum(li))
    return H

h_cnt = [74,80,30,10,6]

print(h_function(h_cnt))

def h_cumulative_function(li):
    N = []
    j = 0
    for i in h_function(li):  # 이게 가능하다!
        j += i
        N.append(round(j,2))
    return N

print(h_cumulative_function(h_cnt))
```

문제53. 한 회사의 영업사원 5명을 모집단이라고 하자. 관심있는 확률변수는 이회사에서의 근무연수로 다음과 같다.

6, 2, 4, 8, 10



```python
# 문제54. 표본의 평균을 파이썬의 forloop문으로 구현하시오
# 모집단에서 추출 가능한 n = 2인 모든 표본과 표본평균

b=[6,2,4,8,10]   # 표본

re = 0
cnt = 0
for x in b:
    for y in b:
        re += ( x + y ) // 2
        cnt += 1
        
print('표본평균의 합:',re, '표본평균의 전체 평균:',re // cnt)
```

※ 표본평균의 전체 평균은 모평균과 같다.

```python
# 문제55. 아래의 표의 표본의 분산을 구하시오

b=[6,2,4,8,10]   # 표본

re = 0
cnt = 0
for x in b:
    for y in b:
        re += (6 - (( x + y ) // 2)) ** 2
        cnt += 1
        
print('표본의 분산:',re // cnt)
```

※ 표본 평균의 분산은 모분산의 표본의 크기(n=2)로 나눈 값이다.

```python
# 문제56. 표본평균의 막대 그래프를 그리기위해 표본 평균의 도수 분포표를 아래와 같이 파이썬으로 출력하시오

b=[6,2,4,8,10]   # 표본

re = []
for x in b:
    for y in b:
        re.append(( x + y ) // 2)

# re.sort()
r_set = set(re)
s = []
num = []
for n in r_set:
#     print(n, re.count(n))
    s.append(n)
    num.append(re.count(n))
```

문제59. 문제9번에서 사용한 영업사원 5명의 근무연수 모집단을 다시 생각해보자



```python
# 표본 분산의 분산값을 파이썬으로 구현하시오

sample = [6, 2, 4, 8, 10]
re = []
for x in sample:
    for y in sample:
        m = ( x + y ) // 2
        re.append((x-m)**2 + (y - m) **2)
        
v_mean = sum(re) // 25
print('표본분산의 평균값:', v_mean)

re2 = 0
for x in re:
    re2 += (x - v_mean)**2

print('표본분산의 분산값:', re2//25)
```
